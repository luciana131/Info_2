```python
class Tarea:
    def __init__(self, descripcion, prioridad):
        self.descripcion = descripcion
        self.prioridad = prioridad
        self.completada = False

    def marcar_completada(self):
        self.completada = True

    @classmethod
    def editar_descripcion(cls, tarea, nueva_descripcion):
        tarea.descripcion = nueva_descripcion

    @classmethod
    def editar_prioridad(cls, tarea, nueva_prioridad):
        tarea.prioridad = nueva_prioridad

    def __str__(self):
        estado = "Completada" if self.completada else "Pendiente"
        return f"Descripción: {self.descripcion} | Prioridad: {self.prioridad} | Estado: {estado}"

class ListaTareas:
    def __init__(self):
        self.tareas = []

    def agregar_tarea(self, tarea):
        self.tareas.append(tarea)

    def ver_tareas_pendientes(self):
        pendientes = [tarea for tarea in self.tareas if not tarea.completada]
        if pendientes:
            print("Tareas Pendientes:")
            for tarea in pendientes:
                print(tarea)
        else:
            print("No hay tareas pendientes.")

    def ver_tareas_completadas(self):
        completadas = [tarea for tarea in self.tareas if tarea.completada]
        if completadas:
            print("Tareas Completadas:")
            for tarea in completadas:
                print(tarea)
        else:
            print("No hay tareas completadas.")

def mostrar_menu():
    print("\nMenú:")
    print("1. Agregar Tarea")
    print("2. Marcar Tarea como Completada")
    print("3. Editar Descripción de Tarea")
    print("4. Editar Prioridad de Tarea")
    print("5. Ver Tareas Pendientes")
    print("6. Ver Tareas Completadas")
    print("7. Salir")

if __name__ == "__main__":
    lista_tareas = ListaTareas()

    while True:
        mostrar_menu()
        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            descripcion = input("Ingrese la descripción de la tarea: ")
            prioridad = input("Ingrese la prioridad de la tarea (Alta/Media/Baja): ")
            tarea_nueva = Tarea(descripcion, prioridad)
            lista_tareas.agregar_tarea(tarea_nueva)
            print("Tarea agregada correctamente.")

        elif opcion == "2":
            tarea = input("Ingrese la descripción de la tarea que desea marcar como completada: ")
            for t in lista_tareas.tareas:
                if t.descripcion == tarea:
                    t.marcar_completada()
                    print("Tarea marcada como completada.")
                    break
            else:
                print("Tarea no encontrada.")

        elif opcion == "3":
            tarea = input("Ingrese la descripción de la tarea que desea editar: ")
            nueva_descripcion = input("Ingrese la nueva descripción: ")
            for t in lista_tareas.tareas:
                if t.descripcion == tarea:
                    Tarea.editar_descripcion(t, nueva_descripcion)
                    print("Descripción de la tarea editada correctamente.")
                    break
            else:
                print("Tarea no encontrada.")

        elif opcion == "4":
            tarea = input("Ingrese la descripción de la tarea cuya prioridad desea editar: ")
            nueva_prioridad = input("Ingrese la nueva prioridad (Alta/Media/Baja): ")
            for t in lista_tareas.tareas:
                if t.descripcion == tarea:
                    Tarea.editar_prioridad(t, nueva_prioridad)
                    print("Prioridad de la tarea editada correctamente.")
                    break
            else:
                print("Tarea no encontrada.")

        elif opcion == "5":
            lista_tareas.ver_tareas_pendientes()

        elif opcion == "6":
            lista_tareas.ver_tareas_completadas()

        elif opcion == "7":
            print("Saliendo del programa...")
            break

        else:
            print("Opción inválida. Por favor, seleccione una opción válida del menú.")
```

```python
class pacientes: #clase padre
    def __init__(self, nombre, edad):
        self.nombre=nombre
        self.edad = edad
        self._resultados= []

    @property
    def resultados (self):
        return self._resultados
    @resultados.setter
    def resultados (self, resultados):
        self._resultados.append(resultados)

    def calcular_promedio(self):
      if not self.resultados:#revisa si hay resultados en la lista
        return 0
         #suma los resultados y los divide por la cantidad  de resultados
      else:
        return sum(self.resultados) / len(self.resultados)

class pacientesNeuro(pacientes): #Clase para calcular los problemas neuro
    def __init__(self, nombre, edad):
      super().__init__(nombre, edad)

    def problemasNeuro(self,limite):
        promedio = self.calcular_promedio()
        if promedio > limite:
            return True
        else:
            return False

class sistemaRegristo:
    def __init__(self):
        self.pacientes=[]

    def registrar(self, paciente):
      if paciente in self.pacientes:
        print('El paciente ya está registrado')
      else:
        self.pacientes.append(paciente)
        print('Paciente agregado')

    def problemasNeurologicos(self, limite):
        if self.pacientes:
            print("Determinando problemas neurológicos...")
            for paciente in self.pacientes:
                promedio = paciente.calcular_promedio()
                if paciente.problemasNeuro(limite):
                    print(f"{paciente.nombre} tiene un problema neurológico.")
                else:
                    print(f"{paciente.nombre} no tiene un problema neurológico.")
        else:
            print("No hay pacientes registrados.")

sistema_registro = sistemaRegristo()
while True:
    print("1. Registrar Paciente")
    print("2. Calcular problemas neurologicos")
    print ("3. Salir")
    menu=input("Ingresar opción:")

    if menu =="1":
      nombre = input("Ingrese el nombre del paciente: ")
      edad = input("Ingresar la edad del paciente: ")
      resultado = float(input("Ingresar el resultado de la prueba neurológica: "))
      nuevo_paciente = pacientesNeuro(nombre, edad)
      nuevo_paciente.resultados = resultado
      sistema_registro.registrar(nuevo_paciente)

    elif menu =="2":
        limite = float(input("Ingrese el valor límite para definir si hay un problema neurológico: "))
        sistema_registro.problemasNeurologicos(limite)
    elif menu =="3":
        print("Menu finalizado")
        break
    else:
        print("opción invalida, vuelva a intentarlo")
```

```python
class Animal:
    def __init__(self, nombre, especie, edad):
        self.nombre = nombre
        self.especie = especie
        self.edad = edad

class Consulta:
    def __init__(self, animal, fecha, diagnostico):
        self.animal = animal
        self.fecha = fecha
        self.diagnostico = diagnostico

class ClinicaVeterinaria:
    def __init__(self):
        self.pacientes = []
        self.consultas = []

    def registrar_paciente(self, paciente):
        if paciente.nombre not in [animal.nombre for animal in self.pacientes]:
            self.pacientes.append(paciente)
            print("Paciente registrado con éxito.")
        else:
            print("Este paciente ya está registrado.")

    def agendar_consulta(self, nombre, fecha, diagnostico):
        paciente_encontrado = False
        for paciente in self.pacientes:
            if paciente.nombre == nombre:
                paciente_encontrado = True
                consulta = Consulta(paciente, fecha, diagnostico)
                self.consultas.append(consulta)
                print("Consulta agendada con éxito.")
                break
        if not paciente_encontrado:
            print("No se encontró al paciente.")

    def mostrar_consultas(self):
        if not self.consultas:
            print("No hay consultas registradas.")
        else:
            print("Consultas registradas:")
            for consulta in self.consultas:
                print(f"Paciente: {consulta.animal.nombre}, Especie: {consulta.animal.especie}, Edad: {consulta.animal.edad}, Fecha: {consulta.fecha}, Diagnóstico: {consulta.diagnostico}")

    def mostrar_menu(self):
        while True:
            print("\n--- Menú de la Clínica Veterinaria ---")
            print("1. Registrar nuevo paciente")
            print("2. Agendar consulta")
            print("3. Mostrar consultas")
            print("4. Salir")
            opcion = input("Seleccione una opción: ")

            if opcion == "1":
                nombre = input("Ingrese el nombre del paciente: ")
                especie = input("Ingrese la especie del paciente: ")
                edad = int(input("Ingrese la edad del paciente: "))
                animal = Animal(nombre, especie, edad)
                self.registrar_paciente(animal)

            elif opcion == "2":
                nombre_paciente = input("Ingrese el nombre del paciente: ")
                fecha = input("Ingrese la fecha de la consulta (DD/MM/AAAA): ")
                diagnostico = input("Ingrese el diagnóstico de la consulta: ")
                self.agendar_consulta(nombre_paciente, fecha, diagnostico)

            elif opcion == "3":
                self.mostrar_consultas()

            elif opcion == "4":
                print("¡Hasta luego!")
                break

            else:
                print("Opción no válida. Por favor, seleccione una opción válida.")


if __name__ == "__main__":
    clinica = ClinicaVeterinaria()
    clinica.mostrar_menu()

https://colab.research.google.com/drive/1i4QDiAd4T3U_W9KtTgfYSq7tCw2vSMJm?usp=sharing#scrollTo=bmwNMMscfHUb
```
